```python
#impor library yang digunakan
import matplotlib.pyplot as plt
import numpy as np
import cv2
import matplotlib.pyplot as plt
import glob

from skimage.color import rgb2gray
from skimage.feature import local_binary_pattern
```

```python
#membuka file pada gambar
img = plt.imread('GambarHp1.jpg')

#membuat fitur histogram warna pada gambar
fig, axs = plt.subplots(1, 3, figsize=(10, 15))
for i, color in enumerate(['red', 'green', 'blue']):
    axs[i].hist(img[..., i].ravel(), bins=256, color=color)
    axs[i].set_xlim([0, 256])
    axs[i].set_ylim([0, 10000])
    axs[i].set_title(color.upper())


#menampilkan histogram warna dari gambar
plt.show()
```
